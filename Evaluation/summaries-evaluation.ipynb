```python
# creating color mappings for consistency
models = ['DeciLM', 'Falcon', 'Gemma', 'Llama 3', 'Mistral', 'Orca 2', 'Zephyr']
magma_palette = sns.color_palette('magma', 7)
model_colors = dict(zip(models, magma_palette))
```

Plot 1: Bertscore Density Graph

```python
bert_df = pd.DataFrame({
    'DeciLM': deci['bert_f1'],
    'Falcon': falcon['bert_f1'],
    'Gemma': gemma['bert_f1'],
    'Llama 3': llama['bert_f1'],
    'Mistral': mistral['bert_f1'],
    'Orca 2': orca['bert_f1'],
    'Zephyr': zephyr['bert_f1']
})

bert_df.head()
```

```python
melted_bert_df = pd.melt(bert_df, var_name="LLM Model", value_name="Bertscore")
melted_bert_df.head()
```

```python
order = melted_bert_df.groupby('LLM Model')['Bertscore'].median().sort_values().index
order
```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(data=melted_bert_df, x='Bertscore', hue='LLM Model', fill=True, common_norm=False, palette="magma")

# Add title and labels
plt.title('Density Plot of Memory Performance Across Models')
plt.xlabel('Bertscore')
plt.ylabel('Density')

# Show the plot
plt.show()
```

Plot 2: Boxplot of Bertscore

```python
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")
sns.boxplot(x = "LLM Model", y = "Bertscore", data = melted_bert_df, order=order, palette='magma', hue="LLM Model")
plt.title('Comparison of Bertscore Across Models', fontsize=16, pad=14)
plt.xlabel('LLM Model')
plt.ylabel('Bertscore')
plt.show()
```

Plot 3: Grid of density plots of each model

```python
# Initialize the FacetGrid object
g = sns.FacetGrid(melted_bert_df, col="LLM Model", col_wrap=7, height=4, hue="LLM Model",  palette="magma")

# Map the seaborn kdeplot function to the FacetGrid
g.map(sns.kdeplot, "Bertscore", fill=True)

# Add titles to the plots
g.set_titles(col_template="{col_name}")

# Show the plot
plt.show()
```
